# Import Modules and Load Dataset

```python
from pycaret.datasets import get_data
diabetes = get_data('diabetes')
```

# Setup and Compare Models

```python
from pycaret.classification import *
clf1 = setup(data = diabetes, target = 'Class variable', numeric_features=['Number of times pregnant'])
```

```python
compare_models(sort='F1', probability_threshold=0.7)
```

```python
best_model = create_model('lr')
```

```python
plot_model(best_model, plot='learning', use_train_data=True)
```

# Finalise and Deploy the Model

```python
import os 
final_lr = finalize_model(best_model)
os.environ["AWS_ACCESS_KEY_ID"] = str("<Your Access Key Here>")
os.environ["AWS_SECRET_ACCESS_KEY"] = str("<Your Secret Access Key Here>")
deploy_model(model = final_lr, model_name = 'lr-for-deployment', platform = 'aws', authentication = {'bucket' : 'mohsinbucket12'}) 
```
